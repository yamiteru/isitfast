```python
import matplotlib.pyplot as plt
import json

dictionary = json.load(open("./stats.json", "r"))

xTrue = [row['i'] for row in dictionary if row['b'] == True]
yTrue = [row['v'] for row in dictionary if row['b'] == True]

xFalse = [row['i'] for row in dictionary if row['b'] == False]
yFalse = [row['v'] for row in dictionary if row['b'] == False]

plt.figure(figsize=(30,20))
plt.scatter(xFalse,yFalse,None,'#d9d9d9')
plt.scatter(xTrue,yTrue,None,'#000000')
plt.show()
```
