```python
import matplotlib.pyplot as plt
import numpy as np
import os

path = "/Users/yamiteru/.isitfast/results"
benchmark = "$for"
mode = "cpu"
pattern = "{}_{}".format(benchmark, mode)
files = os.listdir(path)
colors = {
    'all': '#03fc1c',
    'auto': '#000000',
    'none': '#fc0303'
}

files.sort()

plt.figure(figsize=(30,20))

for file in files:
    if os.path.isfile(os.path.join(path, file)) and pattern in file:
        data = np.fromfile("{}/{}".format(path, file), dtype=np.dtype("uint32"))        
        x = [i for i, key in enumerate(data)]
        y = [key for i, key in enumerate(data)]
        color = "all" if "_all_" in file else "auto" if "_auto_" in file else "none"
        plt.scatter(x,y,None,colors[color],alpha=0.15 if color == "auto" else 0.05)

plt.show()
```
