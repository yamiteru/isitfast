```python
import matplotlib.pyplot as plt
import json

oneJson = json.load(open("./$forEach-cpu-all.json", "r"))
twoJson = json.load(open("./$forEach-cpu-auto.json", "r"))
threeJson = json.load(open("./$forEach-cpu-none.json", "r"))

max = 2000

onex = [i for i, key in enumerate(oneJson)][0:max]
oney = [key for i, key in enumerate(oneJson)][0:max]

twox = [i for i, key in enumerate(twoJson)][0:max]
twoy = [key for i, key in enumerate(twoJson)][0:max]

threex = [i for i, key in enumerate(threeJson)][0:max]
threey = [key for i, key in enumerate(threeJson)][0:max]

plt.figure(figsize=(30,20))
plt.scatter(onex,oney,None,'#347deb') # blue
plt.scatter(twox,twoy,None,'#3aeb34') # green
plt.scatter(threex,threey,None,'#eb4034') # red

plt.show()
```
