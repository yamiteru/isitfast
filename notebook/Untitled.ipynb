```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# plt.ylim(3000, 10000)
# plt.xlim(0, 100)

validator = pd.read_csv("../results/newer/validate.js-$validator.csv")
valibot = pd.read_csv("../results/newer/validate.js-$valibot.csv")
zod = pd.read_csv("../results/newer/validate.js-$zod.csv")

validator = validator.groupby(by="run").rolling(50).median()
valibot = valibot.groupby(by="run").rolling(50).median()
zod = zod.groupby(by="run").rolling(50).median()

sns.lineplot(x="iteration", y="cpu", data=validator)
sns.lineplot(x="iteration", y="cpu", data=valibot)
sns.lineplot(x="iteration", y="cpu", data=zod)
```
