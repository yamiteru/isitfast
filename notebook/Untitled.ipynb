```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.ylim(2000, 7000)
plt.xlim(0, 1500)

baseline = pd.read_csv("/Users/yamiteru/.isitfast/results/baseline.js-$baseline-main.csv")
validator = pd.read_csv("/Users/yamiteru/.isitfast/results/validator.js-$validator-main.csv")
valibot = pd.read_csv("/Users/yamiteru/.isitfast/results/valibot.js-$valibot-main.csv")
zod = pd.read_csv("/Users/yamiteru/.isitfast/results/zod.js-$zod-main.csv")

#baseline = pd.read_csv("/Users/yamiteru/.isitfast/results/baseline.js-$baseline-startup.csv")
#validator = pd.read_csv("/Users/yamiteru/.isitfast/results/validator.js-$validator-startup.csv")
#valibot = pd.read_csv("/Users/yamiteru/.isitfast/results/valibot.js-$valibot-startup.csv")
#zod = pd.read_csv("/Users/yamiteru/.isitfast/results/zod.js-$zod-startup.csv")

baseline = baseline.groupby(by="run").rolling(50).median()
validator = validator.groupby(by="run").rolling(50).median()
valibot = valibot.groupby(by="run").rolling(50).median()
zod = zod.groupby(by="run").rolling(50).median()

sns.lineplot(x="iteration", y="cpu", data=baseline)
sns.lineplot(x="iteration", y="cpu", data=validator)
sns.lineplot(x="iteration", y="cpu", data=valibot)
sns.lineplot(x="iteration", y="cpu", data=zod)

#key = "duration"

#sns.boxplot(data=baseline[key])
#sns.boxplot(data=validator[key])
#sns.boxplot(data=valibot[key])
#sns.boxplot(data=zod[key])
```
